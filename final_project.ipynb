### *2.3. Метрики*

В работе используются следующие метрики: 
* **TTR (Type-Token Ratio)** — коэффициент, измеряющий лексическое разнообразие текста. Рассчитывается как отношение количества уникальных слов (types) к общему числу слов (tokens). Чем выше TTR, тем богаче лексикон автора и тем меньше он прибегает к повторам. Эта метрика особенно важна для сравнения идиостилей, так как позволяет количественно оценить, насколько разнообразна лексика Чехова и Зощенко;
* **MSTTR (Mean Segmental Type-Token Ratio)** — модификация TTR, которая вычисляется как среднее арифметическое TTR на непересекающихся сегментах текста фиксированной длины (например, 1000 слов). Эта метрика менее чувствительна к длине текста, что делает её более надежной для сравнения авторов с разным объемом произведений;
* **Средняя длина слова и предложения** — показатели синтаксической сложности текста. Средняя длина слова дает представление о морфологической структуре (использование длинных, сложных слов vs. коротких). Средняя длина предложения — один из ключевых индикаторов синтаксического стиля: длинные, развернутые предложения часто характерны для «психологической» прозы, а короткие — для разговорной; 
* **Частотный анализ частей речи (POS)** —  используется для оценки морфологической структуры и стилистических особенностей текста. Анализ распределения частей речи (существительные, глаголы, прилагательные, наречия и т.д.) позволяет выявить грамматические предпочтения автора. Например, обилие глаголов может указывать на динамичность повествования, а преобладание существительных и прилагательных — на описательность и психологизм;  
* **Ключевая лексика и биграммы** — выделение наиболее значимых одиночных слов и устойчивых словосочетаний из двух слов для определения главной тематики и контекста исследования. Анализ частотности слов и устойчивых пар позволяет выявить ключевые концепты и темы, характерные для каждого автора, а также определить специфические маркеры его стиля.

### *3.2. Очистка данных*

Сырые тексты содержат ряд элементов, которые не являются художественным текстом и могут исказить результаты анализа: 
* Заголовки и названия рассказов;
* Пустые строки и лишние пробелы;
* Знаки препинания, цифры и спецсимволы;
* Верхний регистр (может привести к тому, что одно и то же слово в разных регистрах будет считаться разным).

Чтобы исключить лишние элементы из текста, была разработана функция `clean_text`, которая: 
1. Удаляет заголовки (первая строка, написанная заглавными буквами или состоит из нескольких слов); 
2. Приводит весь текст к нижнему регистру; 
3. Удаляет все символы, кроме букв русского алфавита и пробелов; 
4. Убирает множественные пробелы, превращая их в один. 

Этот этап критически важен, так как он напрямую влияет на качество последующей токенизации и, следовательно, на все результаты анализа.

```python
# Шаг 1. Функция для очистки текста
def clean_text(text):
    lines = text.split('\n')

    # Удаление пустых строк и названий в текстах
    while lines and lines[0].strip() == '':
        lines.pop(0)

    if lines:
        first_line = lines[0].strip()
        if first_line.isupper() or len(first_line.split()) <= 10:
            lines.pop(0)
    
    text = '\n'.join(lines) # Обратное объединение
    text = text.lower() # Как и всегда: нижний регистр :)
    text = re.sub(r'[^а-яё\s]', '', text) # И тут тоже: удаление всего лишнего
    text = re.sub(r'\s+', ' ', text).strip() # Удаление лишних пробелов

    return text

# Шаг 2. Применение функции
corpus['clean_text'] = corpus['text'].apply(clean_text)
```

### *3.3. Токенизация*

В работе используется токенизатор из библиотеки NLTK (`word_tokenize`), который адаптирован для работы с русским языком. Разбиение на слова является первым и обязательным шагом, без которого невозможно провести дальнейший количественный анализ текста. 

```python
# Шаг 1. Функция для токенизации
def tokenize(text):
    tokens = word_tokenize(text)
    return tokens

# Шаг 2. Применение функции 
corpus['tokens'] = corpus['clean_text'].apply(tokenize)
corpus['token_count'] = corpus['tokens'].apply(len)
```

### *3.4. Лемматизация*

<span style='color: #9C322E;'>Лемматизация</span> — процесс приведения слова к его словарной форме (например, «Чеховым» → «Чехов»). Чтобы привести все формы к единому основанию (лемме), используется лемматизатор на основе библиотеки `pymorhpy3`. 

Лемматизация необходима для анализа по нескольким причинам: 
1. TTR, рассчитанный на леммах, более точно отражает лексическое разнообразие (разные формы одного слова не считаются разными типами);
2. Позволяет корректно анализировать частотность (все формы одного слова приводятся к одной лемме);
3. Устраняет иллюзию большего или меньшего разнообразия из-за использования разных падежей и времен.

Лемматизация учитывает часть речи и сочетания с другими словами, т.е. старается учесть контекст — в отличие от стемминга, который строится на простых правилах (именно поэтому стемминг намного быстрее леммматизации). Она упрощает поиск ключевых слов, анализ текстовых данных и особенно важна при подготовке текстовых данных к кластерному анализу и тематическому моделированию.

```python
# Шаг 1. Словарь для лемматизации
morph = pymorphy3.MorphAnalyzer()


# Шаг 2. Функция для лемматизации
def lemmatize_tokens(tokens):
    lemmas = []
    for token in tokens:
        try:
            lemma = morph.parse(token)[0].normal_form
            lemmas.append(lemma)
        except:
            continue
    return lemmas

# Шаг 3. Применение функции
corpus['lemmas'] = corpus['tokens'].apply(lemmatize_tokens)
corpus['lemma_count'] = corpus['lemmas'].apply(len)
```

```python
# Шаг 4. Вывод результатов
print(f'\033[1mВсего лемм:\033[0m {corpus['lemma_count'].sum()}')
print(f'\033[1mВ среднем на рассказ:\033[0m {corpus['lemma_count'].mean():.0f}')
```

Всего в корпусе 66632 леммы, среднее количество лем на один рассказ составляет 1333. Этот показатель сопоставим со средним количеством токенов на рассказ, что ожидаемо, поскольку лемматизация не изменяет общее количество единиц текста, а лишь приводит их к единой форме. 

## **4. Анализ и вычисления**

### *4.1. TTR и средние значения текста*

Базовые количественные характеристики позволяют получить первое приближение к пониманию стилистических различий между авторами. Каждая из представленных метрик отражает свой уровень языковой организации: лексическое разнообразие (TTR), средняя длина слова — морфологическую структуру, средняя длина предложения — синтаксическую сложность. Объем текстов важен для корректной интерпретации других метрик.

```python
# Шаг 1. Функция для TTR
def calculate_ttr(lemmas):
    if not lemmas:
        return 0
    return len(set(lemmas)) / len(lemmas)

# Шаг 2. Применение функции
corpus['ttr'] = corpus['lemmas'].apply(calculate_ttr)
```

```python
# Шаг 3. Функция для средней длины слова
def calculate_average_word_length(tokens):
    if not tokens:
        return 0
    return sum(len(token) for token in tokens) / len(tokens)

# Шаг 4. Применение функции
corpus['avg_word_len'] = corpus['tokens'].apply(calculate_average_word_length)
```

```python
# Шаг 5. Функция для средней длины предложения
def calculate_average_sentence_length(text):
    sentences = re.split(r'[.!?]+', text)
    sentences = [s.strip() for s in sentences if s.strip()] # Удаление пустых предложений на всякий случай
    if not sentences:
        return 0
    sentence_lengths = [len(s.split()) for s in sentences]
    return sum(sentence_lengths) / len(sentence_lengths)

# Шаг 6. Применение функции
corpus['avg_sentence_len'] = corpus['clean_text'].apply(calculate_average_sentence_length)
```

```python
# Шаг 7. Группировка по авторам и подсчет среднего значения
summary = corpus.groupby('author').agg({
    'ttr': ['mean', 'std'],
    'avg_word_len': ['mean', 'std'],
    'avg_sentence_len': ['mean', 'std'],
    'token_count': ['sum', 'mean']
}).round(3)

# Шаг 8. Наведение красоты в таблице
summary.columns = ['_'.join(col).strip() for col in summary.columns.values]
summary = summary.rename(columns={
    'ttr_mean': 'TTR (среднее)',
    'ttr_std': 'TTR (ст. откл.)',
    'avg_word_len_mean': 'Ср. длина слова',
    'avg_word_len_std': 'Ср. длина слова (ст. откл.)',
    'avg_sentence_len_mean': 'Ср. длина предложения',
    'avg_sentence_len_std': 'Ср. длина предложения (ст. откл.)',
    'token_count_sum': 'Всего слов',
    'token_count_mean': 'Ср. слов на рассказ'
})

# Шаг 9. Вывод результатов
print('\033[1mТаблица «TTR и средние значения»:\033[0m')
summary
```

1. TTR выше у Зощенко (0.460 против 0.424), чувствительна к длине текста. Поскольку рассказы Зощенко почти в три раза короче, его более высокий TTR может быть артефактом объема. Для проверки этого эффекта далее используется MSTTR (раздел 4.5); 
2. Средняя длина слова идентична, что указывает на морфологическую близость текстов; 
3. У Чехова предложения почти в 3 раза длиннее (1923.23 против 741.32), что указывает на принципиально разные синтаксические устройства текстов авторов;
4. Корпус не сбалансирован по объему: Чехов написал почти в 2.6 раза больше текста. Это ограничение учитывается при интерпретации абсолютных частотностей и при выборе метрик (использование MSTTR, нормализация частот). 

### *4.2. Анализ частей речи*

Анализ распределения частей речи (POS-tagging) позволяет оценить морфологическую структуру текстов и выявить грамматические предпочтения авторов. 

```python
# Шаг 1. Задаю анализатор
morph = pymorphy3.MorphAnalyzer()

# Шаг 2. Функция для частей речи (POS)
def get_pos_tags(lemmas):
    '''
    Функция, определяющая части речи для лемм. 
    Возвращает словарь {часть_речи: количество}
    '''
    pos_tags = []

    for word in lemmas:
        try:
            pos = morph.parse(word)[0].tag.POS
            if pos:
                pos_tags.append(pos)
        except:
            continue

    return dict(Counter(pos_tags))

# Шаг 3. Применение функции (на одном рассказе для проверки) 
print('Части речи для рассказа:')
print(get_pos_tags(corpus.iloc[0]['lemmas']))

# Шаг 4. Применение функции (ко всему корпусу)
corpus['pos_counts'] = corpus['lemmas'].apply(get_pos_tags)
print(corpus[['title', 'pos_counts']].head(5))
```

```python
# Шаг 5. Распределение по автоорам
chekhov_lemmas = []
zoschenko_lemmas = []

for _, row in corpus.iterrows():
    if row['author'] == 'Чехов':
        chekhov_lemmas.extend(row['lemmas'])
    else:
        zoschenko_lemmas.extend(row['lemmas'])

# Шаг 6. Применение результатов
chekhov_pos = get_pos_tags(chekhov_lemmas)
zoschenko_pos = get_pos_tags(zoschenko_lemmas)

# Шаг 6. Сводная таблица для результатов
all_pos = sorted(set(chekhov_pos.keys()) | set(zoschenko_pos.keys()))

pos_df = pd.DataFrame({
    'Часть речи': all_pos,
    'Чехов': [chekhov_pos.get(pos, 0) for pos in all_pos],
    'Зощенко': [zoschenko_pos.get(pos, 0) for pos in all_pos]
})

# Добавляем колонку с разницей
pos_df['Разница между авторами'] = pos_df['Чехов'] - pos_df['Зощенко']  # Колонка, показывающая разницу по количеству частей речи между авторами
pos_df['Перевес в сторону'] = pos_df['Разница между авторами'].apply(  # Колонка с информацией, у кого больше частей речи
    lambda x: 'Чехов' if x > 0 else ('Зощенко' if x < 0 else 'Одинаковое количество')
) 


# Шаг 7. Вывод таблицы
print(f'\n\033[1mТаблица «Анализ частей речи»:\033[0m')
pos_df
```

    * **Инфинитивы (INFN):** у Чехова 9484 против 4126 у Зощенко. Используются для выражения безличных, обощенных действий;

### *4.4. Биграммы*

Анализ биграмм (то есть устойчивых комбинаций двух слов) позволяет выявить характерные словосочетания и нарративные маркеры, отличающие стиль одного автора от другого. 

Биграммы в данном случае считаются по словоформам, а не по леммам. Это осознанное решение. 

```python
# Шаг 1. Функция очистки для биграмм (токены)
def clean_tokens_for_bigrams(text):
    """Очистка текста: возвращает токены (без лемматизации)"""
    tokens = word_tokenize(text.lower())
    cleaned = []
    for token in tokens:
        if len(token) > 2 and token not in heavy_stop and token not in names:
            try:
                lemma = morph.parse(token)[0].normal_form
                if lemma not in heavy_stop and lemma not in names:
                    cleaned.append(token)  # ← токен
            except:
                continue
    return cleaned

# Шаг 2. Функция для создания n-gram
def ngrammer(tokens, n=2):
    """Создаёт n-граммы из списка токенов"""
    ngrams = []
    for i in range(len(tokens) - n + 1):
        ngrams.append(tuple(tokens[i:i+n]))
    return ngrams

# Шаг 3. Применяем функцию для токенов
chekhov_text = ' '.join(corpus[corpus['author'] == 'Чехов']['clean_text'])
zoschenko_text = ' '.join(corpus[corpus['author'] == 'Зощенко']['clean_text'])

chekhov_tokens = clean_tokens_for_bigrams(chekhov_text)
zoschenko_tokens = clean_tokens_for_bigrams(zoschenko_text)

# Шаг 4. Создаём биграммы из токенов
chekhov_bigrams = ngrammer(chekhov_tokens, n=2)
zoschenko_bigrams = ngrammer(zoschenko_tokens, n=2)

# Шаг 5. Считаем частоты
chekhov_bigram_freq = Counter(chekhov_bigrams)
zoschenko_bigram_freq = Counter(zoschenko_bigrams)

# Шаг 6. Топ-15 биграмм
chekhov_top = chekhov_bigram_freq.most_common(15)
zoschenko_top = zoschenko_bigram_freq.most_common(15)

# Шаг 7. Создание таблицы
chekhov_bigram_words = [f"{bigram[0]} {bigram[1]}" for bigram, _ in chekhov_top]
chekhov_bigram_freqs = [freq for _, freq in chekhov_top]
zoschenko_bigram_words = [f"{bigram[0]} {bigram[1]}" for bigram, _ in zoschenko_top]
zoschenko_bigram_freqs = [freq for _, freq in zoschenko_top]

bigram_df = pd.DataFrame({
    'Чехов': chekhov_bigram_words,
    'Чехов (абсолютная частота)': chekhov_bigram_freqs,
    'Зощенко': zoschenko_bigram_words,
    'Зощенко (абсолютная частота)': zoschenko_bigram_freqs
})

# Выведение результатов
print('\033[1mТаблица «Топ-15 биграмм»:\033[0m')
bigram_df
```

* У Чехова преобладает биграммы, связанной с темой времени и состояния. Например: «все время», «каждый день», «другой день», «немного погодя», «всякий раз», «всю жизнь». Присутствуют также устойчивые фразы с религиозной семантикой («бога молить», «слава богу», «бог знает»), характерные для русского культурного контекста до XX в. Это может указывать на интерес Чехова к психологическому состоянию героев, их восприятию времени и внутренними переживаниями. 
* Биграммы Зощенко связаны по большей части с диалогом и устной речью. Первое место занимает комбинация «говорит говорит» (12), что может указывать на какие-то тавтологические приемы. Также присутствуют фразы, например: «между прочим», «сказал поп», «или там», «ничего не ответил», «только говорю», «одним словом» — которые также указывают на разговорность рассказов. Вновь встречаются просторечные выражения («сукин сын»), а также характерные для советской эпохи номинации («старый революционер», «городской товарищ»). Это укзаывает на фирменный стиль Зощенко. 

### *4.5. Нормализованные частоты и MSTTR*

Для корректного сравнения лексических характеристик авторов с разным объемом текстов необходимо устранить эффект «длины» корпуса. В этом разделе представлены два подхода к нормализации: частотность на 1000 слов (для ключевой лексики) и MSTTR (для лексического разнообразия). Обе метрики позволяют увидеть реальные различия в лексических предпочтениях, не искаженные разницей в объеме произведений.

#### *4.5.1. Нормализованные частоты ключевой лексики*

```python
# Берём топ-15 из очищенных словарей
chekhov_top15_clean = Counter(chekhov_without_stopwords).most_common(15)
zoschenko_top15_clean = Counter(zoschenko_without_stopwords).most_common(15)

# Общее количество слов (без стоп-слов)
chekhov_total_clean = sum(chekhov_without_stopwords.values())
zoschenko_total_clean = sum(zoschenko_without_stopwords.values())

# Нормализуем на 1000 слов
chekhov_top15_norm = [(word, freq / chekhov_total_clean * 1000) for word, freq in chekhov_top15_clean]
zoschenko_top15_norm = [(word, freq / zoschenko_total_clean * 1000) for word, freq in zoschenko_top15_clean]

# Шаг 2. Сводная таблица с нормализованными частотами
top15_norm_df = pd.DataFrame({
    'Чехов': [word for word, _ in chekhov_top15_norm],
    'Чехов (на 1000 слов)': [round(freq, 2) for _, freq in chekhov_top15_norm],
    'Зощенко': [word for word, _ in zoschenko_top15_norm],
    'Зощенко (на 1000 слов)': [round(freq, 2) for _, freq in zoschenko_top15_norm]
})

print("\033[1m Таблица «Топ-15 слов (на 1000 слов, без стоп-слов)»:\033[0m")
top15_norm_df
```

У Зощенко местоимение «я» (49.68) значительно опережает «он» (12.56), тогда как у Чехова лидирует «он» (30.04), а я на втором месте (29.51). Это подтверждает разницу в нарративной стратегии: Зощенко пишет от первого лица (сказ), Чехов — от третьего (объективное повествование). У Зощенко также заметно выделяются глаголы речи («говорить» — 37.83, «сказать» — 9.10), что говорит о диалогичности его прозы. Присутствие у Зощенко маркеров устной речи («ну», «тут», «конечно») создает эффект спонтанности и живого диалога, тогда как у Чехова такие слова отсутствуют в топе.

#### *4.5.2. MSTTR (Mean Segmental Type-Token Ratio)*

MSTTR — метрика лексического разнообразия, устойчивая к длине текста. Она рассчитывается как среднее арифметическое TTR на непересекающихся сегментах текста фиксированной длины (в нашем случае — 1000 слов). Это позволяет корректно сравнивать авторов с разным объемом произведений, устраняя артефакт, который мы наблюдали в разделе 4.1 (более высокий TTR у Зощенко).

```python
# Шаг 3. Функция для MSTTR
def calculate_msttr(tokens, segment_size=1000):
    """
    Вычисляет MSTTR (Mean Segmental Type-Token Ratio)
    Разбивает текст на сегменты по 1000 слов и усредняет TTR
    """
    if len(tokens) < segment_size:
        return calculate_ttr(tokens)
    
    tt_rs = []
    for i in range(0, len(tokens), segment_size):
        segment = tokens[i:i+segment_size]
        if len(segment) >= 100:
            tt_rs.append(len(set(segment)) / len(segment))
    
    return np.mean(tt_rs) if tt_rs else 0

# Шаг 4. Применяем к корпусу
corpus['msttr'] = corpus['tokens'].apply(calculate_msttr)

# Шаг 5. Сравниваем
msttr_stats = corpus.groupby('author')['msttr'].mean()

# Шаг 6. Создание таблицы
msttr_df = pd.DataFrame(msttr_stats).reset_index()
msttr_df.columns = ['Автор', 'MSTTR']

# Шаг 7. Вывод
print('\033[1mТаблица «MTTR авторов»:\033[0m')
msttr_df
```

MSTTR у Чехова выше, чем у Зощенко (0.606 vs 0.561). Это означает, что на стандартном отрезке текста Чехов использует более разнообразную лексику. 

Более высокий TTR у Зощенко (0.460 vs 0.424), который был продемонстрирован в разделе 4.1, был артефактом меньшего объема его рассказов. MSTTR снимает этот эффект и показывает реальную картину: лексикон Чехова богаче. Это соответствует его репутации мастера психологической прозы, который стремится к точности и разнообразию в выборе слов. Зощенко же сознательно ограничивает свой вокабуляр, имитируя речь «простого» человека — это осознанный художественный прием, а не признак языковой бедности

### *4.6. Применение статистических метрик*

Для проверки статистической значимости различий между авторами использовался U-тест Манна-Уитни — непараметрический критерий для сравнения двух независимых выборок. Этот метод выбран потому, что распределение метрик в выборках не является нормальным (что типично для лингвистических данных). Дополнительно рассчитан Cohen's d — показатель размера эффекта, который позволяет оценить практическую значимость различий (не только статистическую).

```python
# Статистические методы
def cohen_d(x, y):
    """Рассчитывает размер эффекта Коэна для двух выборок"""
    nx, ny = len(x), len(y)
    pooled_std = np.sqrt(((nx-1)*np.var(x) + (ny-1)*np.var(y)) / (nx+ny-2))
    return (np.mean(x) - np.mean(y)) / pooled_std

def compare_authors(metric, use_nonparametric=True):
    """Сравнивает две выборки и возвращает p-value и размер эффекта"""
    chekhov = corpus[corpus['author'] == 'Чехов'][metric].dropna()
    zoschenko = corpus[corpus['author'] == 'Зощенко'][metric].dropna()
    
    if use_nonparametric:
        stat, p = mannwhitneyu(chekhov, zoschenko)
        test_name = "Mann-Whitney U"
    else:
        stat, p = ttest_ind(chekhov, zoschenko)
        test_name = "t-test"
    
    return {
        'metric': metric,
        'test': test_name,
        'p-value': p,
        'significant': p < 0.05,
        'chekhov_mean': chekhov.mean(),
        'zoschenko_mean': zoschenko.mean(),
        'cohen_d': cohen_d(chekhov, zoschenko)
    }

# Применяем ко всем метрикам
metrics_to_test = ['ttr', 'msttr', 'avg_word_len', 'avg_sentence_len']
stats_results = []

for m in metrics_to_test:
    if m in corpus.columns:
        stats_results.append(compare_authors(m))

stats_df = pd.DataFrame(stats_results)

# Компактная версия
compact_df = pd.DataFrame({
    'Метрика': stats_df['metric'],
    'Чехов': stats_df['chekhov_mean'].round(3),
    'Зощенко': stats_df['zoschenko_mean'].round(3),
    'p-value': stats_df['p-value'].round(4),
    "Cohen's d": stats_df['cohen_d'].round(2),
    'Значимо': ['✓' if x else '✘' for x in stats_df['significant']]
})


print("\033[1mТаблица Статистичекое сравнение авторов»\033[0m")
compact_df
```

1. **TTR — не значим.** $p = 0.116 > 0.05$. Более высокий TTR у Зощенко (0.460 / 0.424), который мы видели в разделе 4.1, статистически не значим. Это подтверждает, что он был артефактом меньшего объема рассказов Зощенко. После коррекции с помощью MSTTR (устойчивой к объему) реальная картина становится ясной.
2. **MSTTR (лексическое разнообразие) — статистически значимо.** $p = 0.0005 < 0.05, Cohen's d = 1.17$ (сильный эффект). Это означает, что различия в лексическом разнообразии между авторами не случайны и имеют большую практическую значимость. Чехов на стандартном отрезке текста использует более разнообразную лексику, чем Зощенко.
3. **Средняя длина слова — не значима.** $p = 0.9845$. Различий практически нет, что подтверждает морфологическую близость текстов обоих авторов.
4. **Средняя длина предложения — статистически значима.** $p < 0.0001, Cohen's d = 0.98$ (сильный эффект). Это самое яркое различие между авторами: предложения Чехова почти в 3 раза длиннее, чем у Зощенко. Разница не случайна и отражает принципиально разные нарративные стратегии.

```python
# Коэфиициенты тренда (таблица)

def calculate_trend(data, metric):
    """Рассчитывает коэффициент наклона тренда для метрики"""
    clean_data = data.dropna(subset=['year', metric])
    if len(clean_data) < 3:
        return None, None
    slope, intercept, r_value, p_value, std_err = linregress(clean_data['year'], clean_data[metric])
    return slope, r_value**2

# Считаем тренды для всех метрик
metrics = ['ttr', 'msttr', 'avg_sentence_len']
trends = []

for author in ['Чехов', 'Зощенко']:
    author_data = corpus[corpus['author'] == author]
    for metric in metrics:
        slope, r2 = calculate_trend(author_data, metric)
        if slope is not None:
            direction = "↗" if slope > 0 else "↘"
            trends.append({
                'Автор': author,
                'Метрика': metric,
                'Наклон': round(slope, 4),
                'R²': round(r2, 4),
                'Направление': direction
            })

trend_df = pd.DataFrame(trends)

# Вывод таблицы
print('\033[1mТаблица «Результаты регрессионного анализа метрик идиостиля»:\033[0m')
trend_df
```

Представленная таблица демонстрирует, как ключевые метрики менялись с течением времени для каждого автора. Коэффициент наклона (slope) показывает направление и интенсивность изменения, а R² — долю дисперсии, объясняемую временным фактором.

* **Чехов**: 
    -  Средняя длина предложения демонстрирует наиболее выраженный тренд: наклон 221.8 при R² = 0.44. Это означает, что синтаксис Чехова усложняется к позднему периоду творчества (1900-е годы). Предложения становятся длиннее и структурно сложнее, что соответствует литературоведческим наблюдениям о переходе Чехова от ранней лаконичности к психологической прозе с развернутыми синтаксическими периодами; 
    - TTR показывает умеренное снижение (наклон -0.0106, R² = 0.55), однако MSTTR остается практически неизменным (наклон ≈ 0, R² ≈ 0). Это ключевой результат: падение TTR — артефакт роста объема текстов, а не реальное обеднение лексикона. На стандартном отрезке в 1000 слов лексическое разнообразие Чехова остается стабильным на протяжении всей карьеры;
* **Зощенко**: 
    - Все три метрики у Зощенко показывают слабые и статистически незначимые тренды (R² < 0.07). Это свидетельствует о том, что его стиль сформировался рано и оставался неизменным на протяжении всего исследуемого периода (1923–1936); 
    - Низкое значение R² для TTR (0.048) и MSTTR (0.0005) подтверждает, что лексическая стратегия Зощенко была осознанным художественным приемом, а не результатом временной эволюции. Короткая длина предложения также остается константой, что связано с имитацией устной разговорной речи в сказе.

Разная динамика метрик подтверждает принципиальное различие идиостилей: Чехов эволюционирует в сторону синтаксической сложности, тогда как Зощенко демонстрирует стабильность, обусловленную сознательной стилизацией под «язык маленького человека».


## **6. Результаты**

Проведённое исследование позволило не только подтвердить исходную гипотезу о статистически значимых различиях между идиостилями Чехова и Зощенко, но и уточнить её на уровне отдельных языковых параметров. Использование сбалансированного корпуса (по числу текстов) и метрик, устойчивых к объему (MSTTR), дало возможность избежать артефактов, связанных с разной длиной произведений, и получить достоверные количественные оценки.

### *6.1. Ключевые различия*

1. **Синтаксический уровень** оказался самым контрастным: средняя длина предложения у Чехова почти в 3 раза выше, чем у Зощенко (1924 против 741 слово на рассказ). Статистический тест Манна-Уитни подтвердил значимость этого различия (p < 0.0001), а размер эффекта (Cohen's d = 0.98) указывает на очень сильную разницу. Это говорит о принципиально разных нарративных стратегиях: Чехов тяготеет к сложному, развернутому синтаксису, характерному для психологической прозы; Зощенко, напротив, сознательно имитирует разговорную, «нелитературную» речь, что проявляется в коротких, синтаксически простых предложениях;
2. **Лексический уровень** после коррекции на объем текста (MSTTR) показал, что словарь Чехова на стандартном отрезке в 1000 слов статистически значимо разнообразнее, чем у Зощенко (0.606 против 0.561; p = 0.0005; Cohen's d = 1.17). Этот результат важен, поскольку он опровергает первоначальное впечатление от «сырого» TTR (0.460 против 0.424), которое было артефактом меньшей длины рассказов Зощенко. Более высокий MSTTR Чехова отражает его стремление к точности и разнообразию словоупотребления, тогда как Зощенко, наоборот, сознательно сужает вокабуляр, создавая эффект «языка маленького человека»;
3. **Морфологический уровень** (анализ частей речи) и средняя длина слова (p = 0.9845) не показали значимых различий. Это ожидаемый результат: оба писателя работают в рамках литературной нормы, и их морфологический «скелет» близок.

### *6.2. Качественные различия: ключевая лексика и биграммы*

Анализ ключевой лексики и биграмм выявил содержательные, а не только количественные различия.

1. **Нарративная перспектива:** У Зощенко местоимение «я» занимает первое место по частотности (49.68 на 1000 слов), тогда как у Чехова лидирует «он» (30.04), а «я» — лишь на втором (29.51). Это эмпирически подтверждает ключевое различие: Зощенко пишет от первого лица (сказовая форма), создавая эффект субъективного, личного повествования, тогда как Чехов чаще использует безличное, объективное повествование от третьего лица; 
2. **Диалогичность:** У Зощенко глаголы «говорить» (37.83) и «сказать» (9.10) входят в топ-10 по частотности, что говорит о диалогичности его прозы. У Чехова эти глаголы также присутствуют, но с меньшей относительной частотой, что соответствует его более повествовательной манере;
3. **Социальная семантика:** Уникальная лексика авторов маркирует две разные эпохи и социальные среды. Чехов оперирует словами «господин», «отец», «сестра», «река», «земля» — дворянский мир с присутствием природы. Зощенко же говорит о «гражданине», «кондукторе», «стакане», использует просторечные формы «нету», «дескать» — мир советского городского быта и быта «маленького человека». Биграммы Зощенко («говорит говорит», «между прочим», «сукин сын») подчеркивают разговорность и даже тавтологичность речи его персонажей, тогда как биграммы Чехова («всё время», «каждый день») указывают на его интерес к течению времени и внутренним переживаниям героев.

### *6.3. Динамика идиостилей*

Регрессионный анализ позволил оценить эволюцию стилей во времени:
1. **Чехов:** Обнаружена устойчивая тенденция к усложнению синтаксиса: наклон тренда для средней длины предложения составляет +221.8 слов в год (R² = 0.44). Это соответствует литературоведческим наблюдениям о переходе Чехова от ранней лаконичности к психологической прозе с развернутыми синтаксическими периодами в 1890–1900-е годы. Важно, что MSTTR при этом остается стабильным (тренд ≈ 0), что означает: усложняется не лексика, а именно синтаксическая структура;
2. **Зощенко:** Ни одна из метрик (TTR, MSTTR, длина предложения) не показывает значимых трендов (R² < 0.07). Его стиль, ориентированный на имитацию устной речи, сформировался уже в начале 1920-х и оставался неизменным на протяжении всего исследуемого периода (1923–1936). Это свидетельствует о сознательной и устойчивой стилистической стратегии.

## **7. Ограничения и дальнешие шаги**

### *7.1. Ограничения исследования*

1. **Дисбаланс по объему корпуса.** Несмотря на то, что в работе использовался MSTTR для кореекции, в корпусе сохраняется существенный дисбаланс по общему количеству токенов (48099 у Чехова против 18533 у Зощенко). Это ограничивает возможности сравнительного анализа лексики и абсолютных частотностей. В идеале следовало бы собрать корпус равного объема (например, по 50000 слов на автора), но для рассказов Зощенко это потребовало бы включения значительно большего числа текстов, что могло бы нарушить принцип репрезентативности (включение всех или большинства рассказов); 
2. **Хронологические ограничения.** Тексты Чехова охватывают более широкий временной период (1882–1902), чем тексты Зощенко (1923–1936). Это означает, что выводы о «динамике» стиля Чехова более обоснованы, чем аналогичные выводы для Зощенко, у которого хронологический разброс меньше, а творческий путь короче.
3. **Отсутствие атрибуции метаданных:** Для Зощенко не учитывалось, к какому циклу относится рассказ («Сентиментальные повести», «Рассказы Назара Ильича господина Синебрюхова» и т.д.), а для Чехова — к какому периоду (ранний юмористический, зрелый психологический). Эти подкорпусы могли бы дать более тонкую и интересную картину.

### *7.2. Перпективы для дальнейшего анализа*

Данную работу можно и нужно дополнять. В будущем хотелось бы добавить следующие улучшения: 
1. Во-первых, можно расширить корпус по вертикали, добавив еще больше количество текстов как Чехова, так и Зощенко. Во-вторых, также возможен и горизонтальный рост путем включения в корпус других авторов малой прозы (например, Аркадия Тимофеевича Аверченко или Надежду Александровну Тэффи), что позволило бы поместить Чехова и Зощенко в более широкий литературный контекст и понять, насколько их стилистические особенности уникальны или, напротив, характерны для эпохи; 
2. Можно отдельно проанализировать диалогическую и нарративную речь, сравнив частотность глаголов речи, длину реплик персонажей и авторских ремарок. Это особенно интересно в контексте сказовой прозы Зощенко; 
3. Можно рассчитать не просто среднюю длину предложения, а его структурную сложность: количество придаточных частей, глубину синтаксического дерева, частоту пассивных конструкций. Например, с помозью инструмента `spacy` с моделями для русского.
